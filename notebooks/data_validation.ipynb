# Data Validation
This notebook contains quality checks for the synthetic marketplace dataset.

```python
import pandas as pd

df = pd.read_csv('../data/processed/enrollment_analysis.csv')

def validate_enrollment_data(dataframe: pd.DataFrame) -> dict[str, bool]:
    checks = {
        'no_null_enrollments': dataframe['enrollment_count'].isnull().sum() == 0,
        'positive_premiums': (dataframe['premium_avg'] >= 0).all(),
        'logical_dates': dataframe['month'].between('2024-01', '2024-12').all(),
    }
    return checks

validation_results = validate_enrollment_data(df)
validation_results
```
