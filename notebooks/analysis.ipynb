# Health Equity Analysis for Maine Marketplace
This notebook demonstrates disparity index calculations and a basic visualization workflow.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load processed data
df = pd.read_csv('../data/processed/enrollment_analysis.csv')

# Calculate coverage rates by demographic
coverage_rates = (
    df.groupby(['county', 'race_ethnicity'])
      .apply(lambda x: x['enrollment_count'].sum() / x['eligible_population'].sum())
      .reset_index(name='coverage_rate')
)

# Identify disparities
overall_rate = df['enrollment_count'].sum() / df['eligible_population'].sum()
coverage_rates['disparity_index'] = coverage_rates['coverage_rate'] / overall_rate - 1

# Visualize
plt.figure(figsize=(12, 6))
pivot_table = coverage_rates.pivot(index='county', columns='race_ethnicity', values='disparity_index')
sns.heatmap(pivot_table, annot=True, cmap='RdBu', center=0)
plt.title('Health Coverage Disparity Index by County and Race/Ethnicity')
plt.tight_layout()
```
